```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal
import scipy.fft
import scipy.interpolate
```

```python
x = np.linspace(0, np.pi*6, 200)
y = np.sin(x)# + np.random.normal(0, 1, x.shape)*0.5
```

```python
plt.plot(x, y)
```

### Střední hodnota

```python
med = np.median(y)
print("Střední hodnota: %s" % med)
```

```python
plt.plot(x, y)
plt.plot([x[0], x[-1]], [med, med])
```

### Rozptyl

```python
var = np.var(y)
print("Rozptyl: %s" % var)
```

### Amplituda

```python
amp = np.max(np.abs(y))
print("Amplituda: %s" % amp)
```

```python
plt.plot(x, y)
plt.plot([x[0], x[-1]], [amp, amp])
```

### Trend signálu

```python
z = np.polyfit(x, y, deg=1)
```

```python
z
```

```python
f = np.poly1d(z)
```

```python
f(-6)
```

```python
plt.plot(x, y)
plt.plot(x, f(x))
```

### Energie

```python
energy = np.sum(np.power(y, 2))
```

```python
print("Energie: %s" % energy)
```

```python
power = energy / len(x)
```

```python
print("Výkon: %s" % power)
```
