```python
import numpy as np
import scipy
import scipy.signal
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100, endpoint=False)
```

```python
y = scipy.signal.square(x)
y = np.clip(y, 0, 1)
```

```python
plt.plot(x, y)
```

```python
a = 1
h = lambda t: a*np.e**(-a*t)

kernel = h(x)
kernel = np.clip(kernel, 0, 1)
```

```python
plt.plot(y)
plt.plot(kernel)
```

```python
conv = np.convolve(y, kernel, mode="full")
conv = conv / (np.max(conv) - np.min(conv))
```

```python
plt.plot(y, label="y")
plt.plot(kernel, label="kernel")
plt.plot(conv, label="conv")

plt.legend()
```
