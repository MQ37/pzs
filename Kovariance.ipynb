```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.signal
import seaborn as sns
```

```python
x = np.linspace(-10, 10, 100)

y = 0*x + 1
z = 0*x + 2
```

```python
plt.plot(x, y, label="y")
plt.plot(x, z, label="z")
plt.legend()
```

```python
np.cov(y, z)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)

ysaw = scipy.signal.sawtooth(x)
ysin = np.sin(x)
ysquare = scipy.signal.square(x)
```

```python
plt.plot(x, ysaw, label="saw")
plt.plot(x, ysin, label="sin")
plt.plot(x, ysquare, label="square")
plt.legend()
```

```python
Y = np.stack((x, ysaw, ysin, ysquare))
cov_matrix = np.cov(Y)
```

```python
xlabels = ["X", "SAW", "SIN", "SQUARE"]
ylabels = ["X", "SAW", "SIN", "SQUARE"]

sns.heatmap(cov_matrix, annot=True, xticklabels=xlabels, yticklabels=ylabels)
```
