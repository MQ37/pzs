```python
import numpy as np
import scipy
import scipy.signal
import matplotlib.pyplot as plt
```

```python
T = 100
N = 10000
ERR = 100

x = np.linspace(-N*np.pi, N*np.pi, N*T)
y = np.sin(x) +\
    np.random.normal(scale=ERR, size=(N*T)) * (1+np.arange(N*T))/(T*N*10) 
```

```python
plt.plot(y[:], label="signal with error")

plt.legend()
```

```python
y_matrix = y.reshape(N, T)
```

```python
weights = 1 / (np.arange(N)+1)
wrestored = np.average(y_matrix, axis=0, weights=weights)
restored = y_matrix.mean(axis=0)
```

```python
plt.plot(restored, label="restored")
plt.plot(wrestored, label="restored weights")
plt.plot(np.sin(x[:T]), label="original")

plt.legend()
```
