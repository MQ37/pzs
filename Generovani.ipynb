```python
def saw_signal(N):
    Y = []
    f = 1/2
    for t in range(1, N+1):
        k = np.arange(1, 100, 2)
        sign = (-1) ** (k)
        y = np.sin(2*np.pi*k*f*t) / k
        y = np.sum(y) * (1/2 - 1/np.pi)
        Y.append(y)
    return Y
```

```python
y = saw_signal(10)
```
